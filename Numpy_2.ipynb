```python
import numpy as np
import random as rand
'''
  To find the shape of a numpy array
'''

zero_array = np.array((10))
zero_array
print("Dimension of array is", zero_array.ndim, "& shape of array is", zero_array.shape) # the shape prints out an empty tuple because it's a zero dim array

# to do the same for a zero dimensional array
one_array = np.array([11, 12, 13, 14])
print("Dimension of array one_array is", one_array.ndim, "and the shape of the array is", one_array.shape)

# to do the same for a two dim array, the shape just tells how the matrix is arranged in the array
two_array = np.array(([2,3,3], [3,3,3], [2,2,2]))
print("Dimension of array in two_array is", two_array.ndim, "and the shape of the array is", two_array.shape)

'''
  to do the same for a three dim array amd also find the shape, careful of the number and type of brackets you put the values in. The shape of the array tells you what is in the array, i.e for this 3 dim array,
  (2,2,2) the 1st two is the number of 2d arrays, the 2nd 2 is the number of 2d rows, and the 3rd two is the number of 2d columns
'''
three_array = np.array([[[2,2], [3,3]], [[4,4], [5,5]]])
print ("Dimension of three_array is", three_array.ndim, "and the shape of the array is", three_array.shape)

'''
to do the same for a 4 dim array and tell its shape. Here the shape is (1,3,2,2) 1 is the number of 3d arrays, 3 means in that one 3d array, you have 3 2-d arrays made up of 2 rows(the 3rd number '2'),
and 2 columns (the 4th number 2)
'''
four_array = np.array([[[[2,2],[2,2]],[[2,2],[2,2]],[[2,2],[2,2]]]])
print('The dimension of four_array is', four_array.ndim, "and the shape is", four_array.shape)
```

```python
'''
  RESHAPING AN ARRAY. First the number of elements in a dimension is the shape. Reshaping is done with the object.reshape() method in python. Reshaping allows you to edit the number of elements in a dimension.
  You edit the number of dimensions in an array, inflate i.e from 1 --> 3 dims, or flatten, i.e from 3 --> 1 dims
'''

# here we'll be reshaping the one_array from above
print("one_array currently has", one_array.ndim, "Dimension")

"""
  to reshape the dimensions in the array, howver the matrix you have has to match the number of values in the array to be reshaped, i.e one_array has 4 elements, you cannot reshape into a (2,3),
  array as you need 2 rows of 3 columns, i.e 2*3 values i.e 6 values, however, (2,2) array works since it has 4 values for "2*2, i.e 4 values" for the array, also the array seems to reshape in order,
  i.e first 2 values go in 1st dim and next 2 values go in next dim
"""
one_to_two_array = one_array.reshape(2,2)
one_to_two_array
```

```python
# to test reshaping a (6,) shape array into (2,3) and (3,2) with .reshape() method
new_one_array = np.array ([10,11,12,13,15,18])
new_one_array

new_one_to_two = new_one_array.reshape(2,3)
print("The new_one_to_two array is a (2,3) shape as thus\n",  new_one_to_two)

# To reshape the new_one_to_two array from a (2,3) into a (3,2) array

modified_one_to_two_array = new_one_to_two.reshape(3,2)
print("The modified_one_to_two_array is went from a (2,3) to a (3,2) shape as thus \n", modified_one_to_two_array)

```

```python
'''
  In this case we flatten the array from a 3-dim array, the three_array from above into a one dim or two dim array
'''
print("The three_array is currently a 3-dim array as shown\n", three_array)

# to flatten to a one or 2 dim, i.e chnage it from a (2,2,2) array to a (2,4), (4,2), or a (8,) array, we do so below, here we do so from a 3 to a 2 which is (2,4) first, and then a (4,2)
three_to_two_flat = three_array.reshape(2,4)
print("\n")
print("I have reshaped the three_array from (2,2,2) to (2,4) by flattening it as thus \n", three_to_two_flat)
print("\n", "\n")

# this time we'll change it to a (4,2), printing out the three_arrayy first to show it's shape is still a (2,2,2), and then reshape it to a (4,2)
print("The three array is still a (2,2,2) as shown thus \n", three_array)
print("\n")
modified_one_to_two_array_42 = three_array.reshape(4,2)
print("\n")
print("I have now modified the (2,2,2) array into a (4,2) array as thus \n", modified_one_to_two_array_42)
print("\n", "\n")

# now to change the three_array from (2,2,2) into a one dim array of (8,), we first print the array to show it's (2,2,2) and then flatten it to (8,)
print("The three_array, before flattening to a 1-dim is still a (2,2,2) as thus \n", three_array)
print("\n")
three_array_to_one = three_array.reshape(8) # i used (8,) to flatten, since (8,) is the new shape, but alternatively you can use (-1) to flatten to a one dim array. I prefer using the new shape though
print("\n")
print("The three_array shape (2,2,2) is now flattened to a 1-dim shape (8,) as thus \n", three_array_to_one)

### YYYYYUUUPPPPP!!!!!! it all worked!!!!
```
