```python
# using user defined function and if, else functions to tell an exam taker they have passed or failed their security+ exam by passing their first name, last name and exam scores as arguments
def exam_result_decision(first_name, last_name, exam_score):
    if exam_score > 750:
        print("You have now passed your security+ exam",first_name, last_name, ",", "For your next steps please follow the instructions below")
    else:
        print("Unfortunately you scored below the required pass mark, For your next steps please follow the instructions below")

exam_result_decision("Oluwadamilare", "Oyeleke", 797)
```

```python
# user defined function with conditional statement to require OS update based on current OS
def os_update_check(current_os, newest_os):
    if newest_os - current_os > 0:
        print("Update required, please proceed to the app store to download the latest available update")
    else:
        print("No update is required, OS is up to date")
os_update_check(3,5)
```

```python
# using a return function passed in a variable with a user defined function and conditional statement to determine if os update is needed
def os_update_check(current_os, newest_os):
    return newest_os - current_os
os_check = os_update_check(5,5)
if os_check > 0:
    print("Update required, please proceed to the app store to download the latest available update")
else:
    print("No update is required, OS is up to date")
```

```python
# script using defined function and while loop to print out credit card balance and then showing, your debt is cleared after amount reaches
def credit_card_minimum(minimum_payment, remaining_payment):
  while minimum_payment <= remaining_payment:
    print("You still owe", "$"+str(remaining_payment))
    remaining_payment = remaining_payment - minimum_payment
  print("Your balance has been fully paid off")
credit_card_minimum(55, 1350)
```

```python
# using defined function and while loop to print out Daily sales target, while also showing current total sales
def target_number_sales(total_sales, daily_target):
    while total_sales < daily_target:
        print("Daily target is off by", "$"+str(daily_target - total_sales), "Your current sales are", "$"+str(total_sales), "." )
        total_sales = total_sales + 50
    print("Congratulation!!!, Your total sales goal of", "$"+str(total_sales),   "has been achieved.")
target_number_sales(0, 1500)
```

```python
# using a for loop along with an elif function to print the port in use for certain protocols
for Port in [21, 22, 25, 53, 80, 110, 143, 443, 3306]:
  if Port == 21:
    print('FTP = Port', Port)
  elif Port == 22:
    print ('SSH = Port', Port)
  elif Port == 25:
    print ('SMTP = Port', Port)
  elif Port == 53:
    print('DNS = Port', Port)
  elif Port == 80:
    print('HTTP = Port', Port)
  elif Port == 110:
    print('POP3 = Port', Port)
  elif Port == 143:
    print('IMTP = Port', Port)
  elif Port == 443:
    print('HTTPS = Port', Port)
  elif Port == 3306:
    print('MYSQL = Port', Port)
```

```python
# Script to show how much time available while using the Desktop at my local Library incorporating while loop into a user defined function
def session_period(total_time_available, total_time_used ):
  while total_time_used < total_time_available:
    print("You still have", total_time_available, "minutes left" )
    total_time_available = total_time_available - 60
  print("You no longer have any time left")
session_period(240,0)
```
