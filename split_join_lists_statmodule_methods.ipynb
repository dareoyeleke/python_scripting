```python
# using the .split() function to turn strings into a list with comma as a delimeter
approved_users ="elarson, bmoreno, tshah, sgilmore, eraab"
print ("Before.split():", approved_users)
approved_user = approved_users.split(",")
print("After.split():", approved_user)

# using the .join() function to turn a list into a string
approved_users = ["elarson", "bmoreno", "tshah", "sgilmore", "eraab"]
print("Before.join():", approved_users)
approved_users = ",".join(approved_users)
print("After .join():", approved_users)
```

```python
# using the statistics library to calculate my average tip out for the last 5 shifts as a bartender while joining my average pay converted to a string and the dollar sign to give my average pay a dollar value
import statistics
dares_tipout = [253, 185, 228, 225, 135]
dares_average_tipout = statistics.mean(dares_tipout)
print("Dare's average tip out for the past 5 days is", "$" +str(dares_average_tipout))

# using sorted to oranize dares_tipout from lowest to highest
dares_tipout = [253, 185, 228, 225, 135]
dares_weekone_tipout = sorted(dares_tipout)
print("Dare's tipout from lowest to highest is", dares_weekone_tipout)

# concatenation of Dare's two week tipout and then sorting from lowest to highest before finding the two week average
dares_weektwo_tipout = [228, 400, 103, 287, 316]
dares_twoweek_tipout = dares_weekone_tipout + dares_weektwo_tipout
print("Dare's two week tip out's are", dares_twoweek_tipout)
dares_sorted_twoweek_tipout = sorted(dares_twoweek_tipout)
print ("Dare's two week tipout from lowest to highest is", dares_sorted_twoweek_tipout)
dares_twoweek_avg = statistics.mean(dares_twoweek_tipout)
print ("Dare's average tipout over two weeks is", "$"+str(dares_twoweek_avg))

# using indices to identify dare's lowest and highest tipout
dares_lowest_twoweek_tipout = dares_sorted_twoweek_tipout[0]
print("Dare's lowest tipout in two weeks is", "$"+str(dares_lowest_twoweek_tipout))

dares_highest_twoweek_tipout = dares_sorted_twoweek_tipout[9]
print("Dare's highest tipout in two weeks is", "$"+str(dares_highest_twoweek_tipout))

# using .insert(), .remove() and .append() method's to change Dare's tipout for the shift that he picked up
print("Dare's week two tipout's before extra tipout are", dares_weektwo_tipout)
dares_weektwo_tipout.insert(1, 250) # insert 250 at index 1
print("Dare's week two tipout's after adding his tipout for his extra shift is", dares_weektwo_tipout)
dares_weektwo_tipout.remove(250)
print("Dare's week two tipout was wrong and is currently being edited, therefore his week two tipout is still",dares_weektwo_tipout)
dares_weektwo_tipout.append(350)
print("Dare's new and correct week two tip out is now", dares_weektwo_tipout)
del dares_weektwo_tipout[0:2]
print("There were extra incorrect amounts for Dare's tipout, while we adjust them, his current accurate tipout numbers are", dares_weektwo_tipout)
dares_new_edited_tipout = [340,380]
dares_weektwo_tipout[0:0] = dares_new_edited_tipout
print("Dare's complete accurate week two tipout is now", dares_weektwo_tipout)
```
